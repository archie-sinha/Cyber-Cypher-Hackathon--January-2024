```python
# Visualize random sample images from different classes
def visualize_random_samples(images, labels, class_names, num_samples=5):
    plt.figure(figsize=(15, 8))
    for i, class_name in enumerate(class_names):
        class_indices = np.where(labels == i)[0]
        random_indices = np.random.choice(class_indices, num_samples, replace=False)
        for j, idx in enumerate(random_indices):
            plt.subplot(len(class_names), num_samples, i * num_samples + j + 1)
            plt.imshow(images[idx])
            plt.title(f"{class_name}")
            plt.axis('off')
    plt.show()
```

##Simple Convolutional Neural Network (CNN)

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Define the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))  # Adjust the number of output nodes based on the number of classes

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, Y_train, epochs=50, batch_size=32, validation_data=(X_test, Y_test))
```

```python
# Evaluate the model on the validation set
val_loss, val_accuracy = model.evaluate(X_test, Y_test)
print("Validation Loss:", val_loss)
print("Validation Accuracy:", val_accuracy)
```

```python
# Predict labels for the test set
Y_pred = model.predict(X_test)

# Calculate and report performance metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score
test_accuracy = accuracy_score(Y_test, Y_pred.round())
print("Test Accuracy:", test_accuracy)
```

##Transfer Learning with a Pre-trained Model (e.g., MobileNetV2)

```python
from keras.applications import MobileNetV2
from keras.models import Sequential
from keras.layers import GlobalAveragePooling2D, Dense

# Load MobileNetV2 with pre-trained weights
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Build the model on top of MobileNetV2
model_transferlr = Sequential()
model_transferlr.add(base_model)
model_transferlr.add(GlobalAveragePooling2D())
model_transferlr.add(Dense(512, activation='relu'))
model_transferlr.add(Dropout(0.5))
model_transferlr.add(Dense(3, activation='softmax'))  # Adjust the number of output nodes based on the number of classes

# Compile the model
model_transferlr.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model_transferlr.fit(X_train, Y_train, epochs=10, batch_size=32, validation_data=(X_test, Y_test))
```

```python
# Evaluate the model on the validation set
val_loss, val_accuracy = model_transferlr.evaluate(X_test, Y_test)
print("Validation Loss:", val_loss)
print("Validation Accuracy:", val_accuracy)
```

```python
# Predict labels for the test set
Y_pred = model_transferlr.predict(X_test)

# Calculate and report performance metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score
test_accuracy = accuracy_score(Y_test, Y_pred.round())
print("Test Accuracy:", test_accuracy)
```

#uAgents Functions

```python
import uagents
import asyncio
```

```python
class InventoryAgent(uagents.Agent):
    async def setup(self):
        self.inventory = {}  # Initialize empty inventory

   # @uagents.message_handler
    async def increase_fruit(self, message):
        fruit_name = message.content
        self.inventory[fruit_name] = self.inventory.get(fruit_name, 0) + 1
        print(f"Fruit {fruit_name} increased to {self.inventory[fruit_name]}")

   # @uagents.message_handler
    async def decrease_fruit(self, message):
        fruit_name = message.content
        if fruit_name in self.inventory:
            self.inventory[fruit_name] -= 1
            if self.inventory[fruit_name] == 0:
                del self.inventory[fruit_name]
            print(f"Fruit {fruit_name} decreased to {self.inventory[fruit_name]}")
        else:
            print(f"Fruit {fruit_name} not found in inventory")
```
